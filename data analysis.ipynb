```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
data= movies.merge(credits,on="title")
```

```python
movie = data[["title","id","budget","popularity","genres","keywords","overview","crew","cast","production_countries","revenue","vote_count"]]
```

```python
movie.head()
```

```python
movie.info()
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace=True)
```

```python
movie.isnull().sum()
```

```python
import ast
```

```python
def change(obj):
    t=[]
    for i in ast.literal_eval(obj):
        t.append(i["name"])
    return t 
```

```python
movie["genres"]=movie["genres"].apply(change)
movie["keywords"] = movie["keywords"].apply(change)
```

```python
def change2(obj):
    t=[]
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            t.append(i["name"])
    return t 
```

```python
movie["crew"]=movie["crew"].apply(change2)
```

```python
def new(obj):
    l = []
    imp=0
    for i in ast.literal_eval(obj):
        if imp!=3:
            l.append(i["name"])
        else:
            break
    return l
```

```python
movie["cast"]=movie["cast"].apply(new)
```

```python
movie["overview"]=movie["overview"].apply(lambda x:x.split())
```

```python
movie["genres"]=movie["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movie["keywords"]=movie["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movie["crew"]=movie["crew"].apply(lambda x:[i.replace(" ","") for i in x])
movie["cast"]=movie["cast"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
movie["production_countries"]=movie["production_countries"].apply(change)
```

```python
movie.head()
```

```python
movie["movie_info"]=movie["keywords"]+movie["overview"]+movie["genres"]
```

```python
movie["movie_cast"]= movie["cast"]+movie["crew"]
```

```python
movie.head()
```

```python
movie_cd=movie[["title", "id" ,"budget","popularity","production_countries","revenue","vote_count","movie_info","movie_cast"]]
```

```python
movie_cd.head()
```

```python
movie_cd["movie_info"]=movie_cd["movie_info"].apply(lambda x:" ".join(x))
movie_cd["movie_info"]=movie_cd["movie_info"].apply(lambda x:x.lower())
```

```python
movie_cd.head()
```

```python
movie_cd["movie_cast"]=movie_cd["movie_cast"].apply(lambda x:" ".join(x))
movie_cd["movie_cast"]=movie_cd["movie_cast"].apply(lambda x:x.lower())
```

```python
movie_cd["production_countries"]=movie_cd["production_countries"].apply(lambda x:" ".join(x))
movie_cd["production_countries"]=movie_cd["production_countries"].apply(lambda x:x.lower())
```
